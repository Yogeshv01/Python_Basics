```python
data1= data.groupby(['coustomer_ID', 'Date']).agg(max_brand_count=('Brand count', 'max'),total_brand_count=('Brand count', 'sum')).reset_index()
data1
```

```python
data= data.groupby(['coustomer_ID', 'Date']).agg({"Brand count":['max','sum']},axis=1,inplace=True).reset_index()
print(data)
```
